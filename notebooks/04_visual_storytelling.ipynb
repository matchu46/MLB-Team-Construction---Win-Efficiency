# MLB Team Construction and Win Efficiency

## Executive Summary

This analysis examines how MLB teams convert payroll into wins, with a focus on Wins Above Replacement (WAR) as the primary driver of on-field success.

While payroll influences team performance, results show that spending alone does not guarantee wins. Instead, teams that efficiently convert payroll into WAR consistently outperform expectations, while high-spending teams often experience diminishing returns.

This notebook presents key findings using simplified visuals designed for non-technical stakeholders.

## Key Takeaways

- Spending more does not guarantee success.
- WAR is the strongest predictor of team wins.
- Payroll impacts wins primarily through its ability to generate WAR.
- Efficient roster construction allows lower-payroll teams to remain competitive.
- High payrolls often face diminishing returns without efficient talent allocation.


## Implications for Team Strategy

- Allocate payroll toward high-WAR contributors rather than aggregate spending.
- Emphasize player development and surplus value contracts.
- Monitor cost per WAR as a core performance metric.
- Recognize diminishing returns at extreme payroll levels.


## Conclusion

This analysis demonstrates that competitive success in Major League Baseball
depends less on how much teams spend and more on how effectively they convert
that spending into on-field value.

Organizations that prioritize efficiency consistently outperform financial expectations.
